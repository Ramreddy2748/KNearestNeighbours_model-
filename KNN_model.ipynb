## Load the Data

```python
from sklearn.datasets import load_wine
wine=load_wine()
wine
```

```python
feature_names=wine.feature_names
feature_names
```

## Create a data frame

```python
import pandas as pd
X = pd.DataFrame(wine.data, columns=feature_names)
tar= pd.DataFrame(wine.target, columns=['target'])
X
```

```python
tar
```

```python
wine_df = pd.concat([X, tar], axis=1)
wine_df
```

```python
wine_df.describe()
```

## Visualization on Target variables

```python
import seaborn as sns
sns.countplot(data=tar,x='target')
```

## Split the data for Train and Test

```python
from sklearn.model_selection import train_test_split
X_tr,X_te,y_tr,y_te=train_test_split(X,tar,test_size=0.2,random_state=42)
X_tr
```

## Train data on model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_tr,y_tr)
```

## Predict the outcomes

```python
y_pred=knn.predict(X_te)
y_pred
```

## Caluculate the accuracy of Model

```python
from sklearn.metrics import accuracy_score,precision_score
acc=accuracy_score(y_te,y_pred)
acc
```

## Caluculate for many neighbour_values

```python
import matplotlib.pyplot as plt
l=[]
l1=[]
for i in range(1,7):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_tr,y_tr)
    y_pred=knn.predict(X_te)
    accu=accuracy_score(y_te,y_pred)
    l1.append(i)
    l.append(accu)
l
```

## Visualize data B/w n_neighbors and accuracy

```python
plt.bar(l1,l,color='blue')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy_score')
```

```python
max_acc=max(l)
max_acc
```
